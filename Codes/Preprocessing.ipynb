## Discretization on numerical variables 'price' and 'availability_365'

```python
data3.dtypes
```

```python
Q1=data3['price'].quantile(0.25)
Q2=data3['price'].quantile(0.50)
Q3=data3['price'].quantile(0.75)
IQR=Q3-Q1
uw=Q3+1.5*IQR
lw=Q1-1.5*IQR

column='price'
data3[column]  = np.where((data3[column]<lw) ,'lower_outlier',
                               np.where((data3[column] >=lw) & (data3[column] <=Q1),'low',
                               np.where((data3[column]>Q1) & (data3[column] <=Q2),'low_median',
                               np.where((data3[column]>Q2) & (data3[column] <=Q3),'high_median',
                               np.where((data3[column]>Q3) & (data3[column] <=Q3+1.5*IQR),'high',
                               'upper_outlier')))))
```

```python
data3['price'].value_counts()
```

```python
data3['availability_365'].describe()
```

```python
data3['availability_365'] = np.where((data3['availability_365'] >=0) & (data3['availability_365'] <=90),'Fully_booked',
                                np.where((data3['availability_365'] >90) & (data3['availability_365'] <=180),'mostly_booked',
                               np.where((data3['availability_365']>180) & (data3['availability_365'] <=270),'frequently_booked',
                                np.where((data3['availability_365']>270) & (data3['availability_365']<=360), 'rarely_booked',
                                                                       'no_booking'))))
```
